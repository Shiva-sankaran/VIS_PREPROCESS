```python
def extract_first_numerical(s):
    match = re.search(r'(\d+\.\d+)', s)
    if match:
        return float(match.group())
    else:
        return None

# Apply the function to the column
final_df['Road traffic death rate'] = final_df['Road traffic death rate'].apply(extract_first_numerical)
```
